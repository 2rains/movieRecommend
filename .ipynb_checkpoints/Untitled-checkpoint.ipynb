```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
#%%
urls = 'C:\Iot1500\documents_movie'
#%%
driver = webdriver.Chrome('C:\\Iot1500\\chromedriver.exe') # 드라이버 잡기
driver.get('https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=17170&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=')
```

```python
# 네이버 무비 노트북 평점 항목
#/html/body/div/div/div[5]/ul/li[7]/div[2]/dl/dt/em[1]/a/span
from selenium.webdriver.common.by import By

driver.find_element(By.XPATH,'//*[@id="actualYnLable"]').click() # 관람객 평점으로 클릭
driver.implicitly_wait(1) # 1초 대기
total_lst = []
for v in range(5):
    for i in range(10):
        time.sleep(.4)
        # 관람객을 클릭했을때 나타나는 영화리스트
        watcher = driver.find_element(By.XPATH,'/html/body/div/div/div[5]/ul/li[{num}]/div[2]/dl/dt/em[1]/a/span'.format(num=i+1))
        watcher.click()
        time.sleep(0.4)
        page = driver.page_source
        driver.back()
        info = bs(page,'lxml')
        td_list = info.find_all('td','title')
        lst = []
        for i in range(len(td_list)):
            lst.append(td_list[i].a.string)
            total_lst.append(lst)
            time.sleep(0.4)
            
    time.sleep(.4)
    next_page = driver.find_element(By.XPATH,'//*[@id="pagerTagAnchor{num2}"]/span'.format(num2 = v+2))
    time.sleep(.4)
    next_page.click()
print(total_lst)
```

```python
total_lst.count(total_lst[0])
# ['부산행', '맨 인 더 다크', '노트북', '아수라']
```

```python
#%% 영화 댓글을 쓴 횟수가 1 이상이어서 두번 이상 반복되는 자료는 삭제함.
for i in range(len(total_lst)):
    if total_lst.count(total_lst[i]) != 1:
        try:
            while(total_lst.count(total_lst[i]) != 1):
                total_lst.remove(total_lst[i])
        except:
            continue
#%%
print(total_lst)
```
